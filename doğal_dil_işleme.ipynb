```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("gender_classifier.csv", encoding="latin-1")
data.head(3)
```

```python
data.columns
```

```python
data1 = pd.concat([data.gender,data.description], axis=1)
```

```python
data1
```

```python
data1.dropna(inplace=True)
data1
data1.reset_index(drop=True , inplace=True)
```

```python
data1.gender = [0 if i=="female" else 1 for i in  data1.gender]
data1.head(3)
```

```python
import re
metin = re.sub("[^a-zA-Z]", " ",data1.description[0])
harfler = metin.lower()
bol=harfler.split()
```

```python
!pip install nltk
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
stop = nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
metin = [ps.stem(i) for i in bol if not i in set(stopwords.words("english"))]
metinson = " ".join(metin)
```

```python
liste = []
for j in range(1000):
  metin = re.sub("[^a-zA-Z]" ," ", data1.description[j])
  metin = metin.lower()
  metin = metin.split()
  metin = [ps.stem(i) for i in metin if not i in set(stopwords.words("english")) ]
  metinson = " ".join(metin)
  liste.append(metinson)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000)
x = cv.fit_transform(liste).toarray()
y = data1.iloc[:1000,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x,y, random_state=63 , test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
model = gn.fit(x_train, y_train)
model.score(x_test,y_test)
model.predict(x_test)
```
