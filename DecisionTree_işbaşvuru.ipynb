```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DecisionTreesClassificationDataSet.csv")
df.head()
```

```python
df.info()
```

```python
dummy = pd.get_dummies(df[["SuanCalisiyor?", "Top10 Universite?", "StajBizdeYaptimi?", "IseAlindi"]] ,drop_first=1)
```

```python
dummy.head()
```

```python
df = pd.concat([dummy, df], axis=1)
```

```python
df.head()
```

```python
df = df.drop(columns=["SuanCalisiyor?", "Top10 Universite?", "StajBizdeYaptimi?", "IseAlindi"])
```

```python
df.head()
```

```python
df["Egitim Seviyesi"].unique()
```

```python
dm = pd.get_dummies(df["Egitim Seviyesi"], drop_first=1)
```

```python
df = pd.concat([dm, df], axis=1)
```

```python
df = df.drop(columns=["Egitim Seviyesi"])
```

```python
df.head()
```

```python
df.groupby('SuanCalisiyor?_Y')['Deneyim Yili', 'StajBizdeYaptimi?_Y'].mean()
```

```python
x = df.drop(["IseAlindi_Y"],axis=1)
y = df["IseAlindi_Y"]
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
model = dtc.fit(x,y)
```

```python
df.head(1)
```

```python
model.predict([[0,1,0,1,0,5,4]])
```
